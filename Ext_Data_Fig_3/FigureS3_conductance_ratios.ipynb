```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fig1():
    
    fig = plt.figure(figsize = (4,3))
    ax = fig.add_subplot(111)
    
    gleak = np.arange(0,2,0.001)
    
    ngi = np.array([0.1,0.5,1.0,2.0])

    syn_ratio_VS = 8.64
    idf_ratio = 2 # ratio of excitatory and inhibitory ionic driving forces
    
    for i in range(4):
    
        r = gleak / (syn_ratio_VS * ngi[i] + idf_ratio * syn_ratio_VS * gleak)
        plt.plot(gleak, r, label = 'n * gi = ' + np.str(ngi[i]) + ' gleak')
    
    plt.legend(loc = 4, frameon = False)
    plt.xlabel('gleak')
    plt.ylabel('synaptic conductance ratio ge/gi')
    
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines['left'].set_position(('outward', 10))
    ax.spines['bottom'].set_position(('outward', 10))

    plt.xlim([0,2])
    plt.ylim([0,0.06])
    
fig1()
```

```python
def fig2():
    
    fig = plt.figure(figsize = (4,3))
    ax = fig.add_subplot(111)
    
    x = np.arange(0, 2, 0.001)
    
    syn_ratio_LPi43 = 17.59 # ratio of excitatory to inhibitory synapses
    syn_ratio_LPi34 = 18.91
    syn_ratio_VS = 8.64
    
    idf_ratio = 2 # ratio of excitatory and inhibitory ionic driving forces
    
    g_ratio_LPi43 = 1 / (syn_ratio_LPi43 * (x + idf_ratio))
    g_ratio_LPi34 = 1 / (syn_ratio_LPi34 * (x + idf_ratio))
    g_ratio_VS = 1 / (syn_ratio_VS * (x + idf_ratio))

    plt.plot(x, g_ratio_LPi43, color = 'C0')
    plt.plot(x, g_ratio_LPi34, color = 'C1')
    plt.plot(x, g_ratio_VS, color = 'C2')
    
    plt.ylabel('synaptic conductance ratio $g_{e}$ / $g_{i}$')
    plt.xlabel('total conductance ratio n * $g_{i}$ / $g_{leak}$')
    
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines['left'].set_position(('outward', 10))
    ax.spines['bottom'].set_position(('outward', 10))
    
    plt.xlim([0,2])
    plt.ylim([0,0.06])
    
fig2()
```

```python
def fig3():
    
    fig = plt.figure(figsize = (3,2.3))
    ax = fig.add_subplot(111)
    
    x = np.arange(0, 2, 0.001)
    syn_ratio_LPi43 = 17.59 # ratio of excitatory to inhibitory synapses
    syn_ratio_LPi34 = 18.91
    syn_ratio_VS = 8.64
    
    idf_ratio = 2 # ratio of excitatory and inhibitory ionic driving forces
    
    g_ratio_LPi43 = syn_ratio_LPi43 * (x + idf_ratio)
    g_ratio_LPi34 = syn_ratio_LPi34 * (x + idf_ratio)
    g_ratio_VS = syn_ratio_VS * (x + idf_ratio)

    plt.plot(x, g_ratio_LPi34, color = 'C1', label = 'LPi3-4', linewidth=1.5)
    plt.plot(x, g_ratio_LPi43, color = 'C0', label = 'LPi4-3', linewidth=1.5)
    plt.plot(x, g_ratio_VS, color = 'C2', label = 'VS', linewidth=1.5)
    
    
    plt.ylabel('syn. conductance ratio ($g_{i}$/$g_{e}$)', size=10)
    plt.xlabel('total conductance ratio ($n_{i}$*$g_{i}$/$g_{leak}$)', size=10)
    ax.set_xticks([0, 0.5, 1.0, 1.5, 2.0])
    ax.set_xticklabels(('0', '0.5', '1.0', '1.5', '2.0'), size=10)
    ax.set_yticks([0, 20, 40, 60, 80])
    ax.set_yticklabels(('0', '20', '40', '60', '80'), size=10)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines['left'].set_position(('outward', 7))
    ax.spines['bottom'].set_position(('outward', 7))
    #ax.legend(loc = 4, frameon = False)
    plt.xlim([0,2])
    plt.ylim([0,80])
    
fig3()

bbox_inches = 'tight'
plt.savefig('C:\\Users\\gammer\\Desktop\\DATA Surface\\LPi Opponency\\LPi Manuscript\\FigureS3_Conductance_ratios\\\
FigS3_ge_gi_conductance_ratio.pdf',bbox_inches='tight', dpi=600, transparent=True)
plt.show()
```
