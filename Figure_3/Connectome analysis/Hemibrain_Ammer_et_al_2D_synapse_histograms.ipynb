```python
def plot_synapses_3d(list_df_synapses, **kwargs):
    kwarguments = {'xlims': (-5000, 25000),
               'ylims': (16000, 38000),
               'zlims': (20000, 36000),
               'azim': 40}
    kwarguments.update(kwargs)

    fig = plt.figure(figsize=(24,8))    
    for i, df_synapses in enumerate(list_df_synapses):
        ax = plt.subplot(1,3,i+1, projection='3d', proj_type = 'ortho')
        df_synapses_pre = df_synapses[df_synapses.type=='pre']
        df_synapses_post = df_synapses[df_synapses.type=='post']

        ax.scatter(df_synapses_pre.x, df_synapses_pre.y, df_synapses_pre.z, color='r', s=4, alpha=0.3)
        ax.scatter(df_synapses_post.x, df_synapses_post.y, df_synapses_post.z, color='c', s=4, alpha=0.3)
        
        ax.set_xlim(kwarguments['xlims'])
        ax.set_ylim(kwarguments['ylims'])
        ax.set_zlim(kwarguments['zlims'])
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        
        ax.view_init(elev=30, azim=kwarguments['azim']) #

xlims = (-3000, 7000)
ylims = (5000, 15000)
zlims = (20000, 30000)

plot_synapses_3d([LPi34_synapses, LPi43_synapses], xlim=xlims, ylims=ylims, zlims=zlims, azim=170)
```
