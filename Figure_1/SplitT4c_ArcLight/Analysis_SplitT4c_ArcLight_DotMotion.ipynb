```python
# Python_version 2.7.15

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
import os
import scipy.io
from scipy import stats
%matplotlib inline
from scipy.signal import savgol_filter
import octopus as oct
```

```python
# Import Data for T4 Axon Terminals

filenames =['SplitT4c_ArcLight_TM4_8dir_002_2021-02-03_12.15_Ftrace_mean',
            'SplitT4c_ArcLight_TM8dir_5s_011_2021-02-18_17.56_Ftrace_mean',
            'SplitT4c_ArcLight_TM8dir_5s_002_2021-02-19_11.15_Ftrace_mean',
            'SplitT4c_ArcLight_TM8dir_5s_009_2021-02-19_16.16_Ftrace_mean',
            'SplitT4c_ArcLight_TM4_alldir_005_2021-02-23_12.09_Ftrace_mean',
            'SplitT4c_ArcLight_TM4_TM4_8dir_009_2021-02-23_15.58_Ftrace_mean',           
            'SplitT4c_ArcLight_TM4_8dir_005_2021-02-24_12.46_Ftrace_mean',
            'SplitT4c_ArcLight_TM4_alldir_004_2021-04-22_14.46TM4_alldir_axons',
            'SplitT4c_ArcLight_TM4_alldir_010_2021-04-22_17.19TM4_alldir_axons',
            'SplitT4c_ArcLight_TM4_8dir_007_2021-05-19_16.47_Ftrace_mean',
            'SplitT4c_ArcLight_TM4_8dir_002_2021-05-21_11.46_Ftrace_mean',
            '2022_04_27_SplitT4c_Arclight_TM8dir_001_2022-04-27_16.32_Ftrace_mean_axons',
            '2022_04_27_SplitT4c_Arclight_TM8dir_003_2022-04-27_17.30_Ftrace_mean_axons']
```

```python
# Import Data
time = np.load('time.npy')

for i in range(len(filenames)):
    a = np.load(filenames[i] + '.npy')
    if(i==0):
        Ftrace_all = a
    if(i==1):
        Ftrace_all = np.stack((Ftrace_all,a), axis=2)   
    if(i>1):
        a = a[:,:,np.newaxis]
        Ftrace_all = np.concatenate((Ftrace_all,a), axis=2)   
```

```python
fig = plt.figure(figsize = (4, 2.6))

plt.subplots_adjust(wspace=0.1)

ax = fig.add_subplot(121)
#ax.axhline(y=0,xmin=0.0001,xmax=0.9999, color='0.05', linestyle=':', linewidth = 1.5)
plt.plot(time,savgol_filter(Ftrace_all[:,5,:],9,1,axis=0), color = 'C3', alpha = 0.35, linewidth = 1)
plt.plot(time,savgol_filter(np.mean(Ftrace_all[:,5,:], axis=1),3,1), color = 'C3', linewidth = 2.5)
plt.axvspan(0, 5, facecolor='0.2', alpha=0.2)
plt.xlim([-3,7.5])
plt.ylim([-0.05,0.09])
plt.axis('off')

ax = fig.add_subplot(122)
#ax.axhline(y=0,xmin=0.0001,xmax=0.9999, color='0.05', linestyle=':', linewidth = 1.5)
plt.plot(time,savgol_filter(Ftrace_all[:,1,:],9,1,axis=0), color = 'C3', alpha = 0.35, linewidth = 1)
plt.plot(time,savgol_filter(np.mean(Ftrace_all[:,1,:], axis=1),3,1), color = 'C3', linewidth = 2.5)
plt.axvspan(0, 5, facecolor='0.2', alpha=0.2)
plt.xlim([-3,7.5])
plt.ylim([-0.05,0.09])
plt.axis('off')

bbox_inches = 'tight'
#plt.savefig('Fig1_SplitT4c_Ftraces.pdf',bbox_inches='tight', dpi=600, transparent=True)
plt.show()
```

```python
rotations = [0,45,90,135,180,225,270,315,360]
Ftrace_mean = np.zeros([len(rotations),len(filenames)])
Ftrace_sem = np.zeros([len(rotations)])
for i in range(len(rotations)):
    Ftrace_sem[i] = stats.sem(np.mean(Ftrace_all[63:145,i,:],0))
    for j in range(len(filenames)):
        Ftrace_mean[i,j] = np.mean(Ftrace_all[np.int(np.where(time == 0)[0]):145,i,j])
```

```python
fig = plt.figure(figsize = (2.3, 2.3))

ax = fig.add_subplot(111) #PLOT directional tuning

ax.axhline(y=0,xmin=0.0001,xmax=0.9999, color='0.05', linestyle=':', linewidth = 1)
plt.plot(rotations, Ftrace_mean, color='C3', linewidth = 1, zorder = 0, alpha=0.35)
plt.plot(rotations, np.mean(Ftrace_mean,1), linewidth = 2.5, color = 'C3')
plt.errorbar(rotations, np.mean(Ftrace_mean,1), yerr = Ftrace_sem, linewidth = 2, color = 'C3')

plt.xticks(rotations)
ax.set_xticklabels(rotations)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines['left'].set_position(('outward', 7))
ax.spines['bottom'].set_position(('outward', 7))
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(0.5)
ax.set_xlabel(u'Direction [°]', fontsize = 10)
ax.set_ylabel('-$\Delta F/F$ (%)', fontsize = 10)
yticks = np.array([-0.1,-0.05,0,0.05,0.1])
ax.set_yticks(yticks)
ax.set_yticklabels(('-10','-5','0','5','10'), size=10)
xticks = np.array([0,90,180,270,360])
ax.set_xticklabels(xticks, size=11)
ax.set_xticks(xticks)
plt.ylim([-0.055,0.055])

bbox_inches = 'tight'
#plt.savefig('Fig1_SplitT4c_Tuning.pdf',bbox_inches='tight', dpi=600, transparent=True)
plt.show()
```

```python
###### Input to MOI function
rot_rad = np.radians(rotations[0:8])
radius = np.mean(Ftrace_mean[0:8,:],1)
```

```python
MOI_flies = np.empty(len(Ftrace_mean[0,:]))
for i in range(len(Ftrace_mean[0,:])):
    MOI_flies[i] = oct.MOI(rot_rad, Ftrace_mean[0:8,i])
```

```python
LDir_flies = np.empty(len(Ftrace_mean[0,:]))
for i in range(len(Ftrace_mean[0,:])):
    LDir_flies[i] = oct.LDir(rot_rad, Ftrace_mean[0:8,i])
```

```python
R_deg_flies = np.empty(len(Ftrace_mean[0,:]))
for i in range(len(Ftrace_mean[0,:])):
    R_deg_flies[i] = oct.R_deg(rot_rad, Ftrace_mean[0:8,i])
```

```python
# Save Ftrace as numpy array

np.save('SplitT4c_Ftrace_mean.npy', Ftrace_mean)
np.save('SplitT4c_Ftrace_all.npy', Ftrace_all)
np.save('SplitT4c_MOI.npy', MOI_flies)
np.save('SplitT4c_LDir.npy', LDir_flies)
np.save('SplitT4c_R_deg.npy', R_deg_flies)
```
