```python
# Calculate timepoint at which impulse response reaches value of t_value (0.5)

t_value = 0.5
min_time = 950 # minimal time that should be taken into consideration (10 - 950/100 corresponds to 0.5s)

t_half_R24E09 = np.empty(len(imp_resp_R24E09[0,:]))
for i in range(len(imp_resp_R24E09[0,:])):
    f = interpolate.interp1d(np.arange(-0,11), imp_resp_R24E09[:,i])
    xnew = np.arange(0, 10, 0.01)
    ynew = f(xnew)   # use interpolation function returned by `interp1d`
    index = np.argmin(np.abs(ynew[0:min_time] - t_value))
    ind = np.argwhere(ynew[0:min_time] == ynew[index])
    ind = ind.astype(np.float32)
    t_half_R24E09[i] = 10 - ind/100
    
t_half_R20D01 = np.empty(len(imp_resp_R20D01[0,:]))
for i in range(len(imp_resp_R20D01[0,:])):
    f = interpolate.interp1d(np.arange(-0,11), imp_resp_R20D01[:,i])
    xnew = np.arange(0, 10, 0.01)
    ynew = f(xnew)   # use interpolation function returned by `interp1d`
    index = np.argmin(np.abs(ynew[0:min_time] - t_value))
    ind = np.argwhere(ynew[0:min_time] == ynew[index])# here is your result
    ind = ind.astype(np.float32)
    t_half_R20D01[i] = 10 - ind/100
    
t_half_R38G02 = np.empty(len(imp_resp_R38G02[0,:]))
for i in range(len(imp_resp_R38G02[0,:])):
    f = interpolate.interp1d(np.arange(-0,11), imp_resp_R38G02[:,i])
    xnew = np.arange(0, 10, 0.01)
    ynew = f(xnew)   # use interpolation function returned by `interp1d`
    index = np.argmin(np.abs(ynew[0:min_time] - t_value))
    ind = np.argwhere(ynew[0:min_time] == ynew[index])# here is your result
    ind = ind.astype(np.float32)
    t_half_R38G02[i] = 10 - ind/100
    
t_half_VT50384 = np.empty(len(imp_resp_VT50384[0,:]))
for i in range(len(imp_resp_VT50384[0,:])):
    f = interpolate.interp1d(np.arange(-0,11), imp_resp_VT50384[:,i])
    xnew = np.arange(0, 10, 0.01)
    ynew = f(xnew)   # use interpolation function returned by `interp1d`
    index = np.argmin(np.abs(ynew[0:min_time] - t_value))
    ind = np.argwhere(ynew[0:min_time] == ynew[index])# here is your result
    ind = ind.astype(np.float32)
    t_half_VT50384[i] = 10 - ind/100
```

```python
# Make Scatter Plot Tau_half

pos = [0,1,2,3]

fig = plt.figure(figsize = (3,3))
#plt.subplots_adjust(wspace = 0.5)
ax = fig.add_subplot(111)
w = 0.6
wb = 0.6
data = [t_half_R24E09, t_half_R20D01, t_half_R38G02, t_half_VT50384]
colors = ['C2', 'C1', 'C0', 'C3']
alpha = [0.35, 0.15, 0.15, 0.15] 
ax.axhline(y=0,xmin=0.0001,xmax=0.9999, color='0.05', linestyle=':', linewidth = 1)

for j in range (len(data)):
    ax.plot([pos[j]-w/2, pos[j]+w/2],[np.mean(data[j]),np.mean(data[j])],
            color=colors[j], linestyle='-', linewidth = 4)
    ax.errorbar(pos[j],np.mean(data[j]),yerr=[stats.sem(data[j])], color='k',zorder = 0)    
    for i in range(len(data[j])):
        ax.scatter(pos[j] + np.random.random(1) * w/2 - w/4, data[j][i], s=25, color=colors[j], alpha=alpha[j], zorder = 2) 
   
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines['left'].set_position(('outward', 10))
ax.spines['bottom'].set_position(('outward', 10))
ax.set_xticks(pos)
ax.set_xticklabels(('VS',' LPi \n3-4', ' LPi \n4-3', 'T4T5c'), size=11)
ax.set_ylabel(r'$\tau$$_{1/2}$ [s]', size=11)
ax.tick_params(axis='both', which='major', labelsize=11)
plt.ylim([0,3])

bbox_inches = 'tight'
#plt.savefig('FigS_MC_RF_Tau_half_boxplots.pdf',bbox_inches='tight', dpi=600, transparent=True)

plt.show()
```
