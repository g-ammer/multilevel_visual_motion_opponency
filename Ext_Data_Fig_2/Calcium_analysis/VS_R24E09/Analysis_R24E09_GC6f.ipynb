```python
filenames =['R24E09_GC6f001_2020-07-02_17.34_1s3p_Ftrace_mean',
            'R24E09_GC6f007_2020-07-03_16.35_1s3p_Ftrace_mean',
            'R24E09_GC6f008_2020-07-03_17.13_1s3p_Ftrace_mean',
            'R24E09_GC6f003_2020-07-06_12.34_1s3p_Ftrace_mean',
            'R24E09_GC6f007_2020-07-06_15.23_1s3p_Ftrace_mean',
            'R24E09_GC6f009_2020-07-06_16.12_1s3p_Ftrace_mean',
            'R24E09_GC6f003_2020-07-07_11.18_1s3p_Ftrace_mean',
            'R24E09_GC6f007_2020-07-07_13.35_1s3p_Ftrace_mean',
            'R24E09_GC6f003_2020-07-07_19.40_1s3p_Ftrace_mean',
            'R24E09_GC6f001_2020-07-10_10.43_1s3p_Ftrace_mean',
            'R24E09_GC6f_SINGalldir_005_2021-05-19_15.36_Ftrace_mean_SINGalldir']
```

```python
# Import Data
time = np.load('time_1s3p.npy')

for i in range(len(filenames)):
    a = np.load(filenames[i] + '.npy')
    if(i==0):
        Ftrace_all = a
    if(i==1):
        Ftrace_all = np.stack((Ftrace_all,a), axis=2)   
    if(i>1):
        a = a[:,:,np.newaxis]
        Ftrace_all = np.concatenate((Ftrace_all,a), axis=2)   
```

```python
fig = plt.figure(figsize = (5,3))

ax = fig.add_subplot(121)
plt.plot(time,Ftrace_all[:,6,:], color = 'C2', alpha = 0.35, linewidth = 1.5)
plt.plot(time,np.mean(Ftrace_all[:,6,:], axis=1), color = 'C2', linewidth = 3.5)
plt.axvspan(0, 1, facecolor='0.2', alpha=0.2)
plt.xlim([-1.5,2.5])
plt.ylim([-1.5,3.8])
plt.axis('off') 

ax = fig.add_subplot(122)
plt.plot(time,Ftrace_all[:,2,:], color = 'C2', alpha = 0.35, linewidth = 1.5)
plt.plot(time,np.mean(Ftrace_all[:,2,:], axis=1), color = 'C2', linewidth = 3.5)
plt.axvspan(0, 1, facecolor='0.2', alpha=0.2)
plt.xlim([-1.5,2.5])
plt.ylim([-1.5,3.8])
plt.axis('off')

bbox_inches = 'tight'
#plt.savefig('C:\\Users\\gammer\\Desktop\\DATA Surface\\LPi Opponency\\plots_LPi_ms\\
#FigS2_R24E09_Ftraces.pdf',bbox_inches='tight', dpi=600, transparent=True)
plt.show()
```

```python
rotations = [0,45,90,135,180,225,270,315,360]
Ftrace_mean = np.zeros([len(rotations),len(filenames)])
Ftrace_sem = np.zeros([len(rotations)])
for i in range(len(rotations)):
    Ftrace_sem[i] = stats.sem(np.mean(Ftrace_all[38:63,i,:],0))
    for j in range(len(filenames)):
        Ftrace_mean[i,j] = np.mean(Ftrace_all[np.int(np.where(time == 0)[0]):63,i,j])
```

```python
fig = plt.figure(figsize = (2.3, 2.3))


ax = fig.add_subplot(111) #PLOT directional tuning

ax.axhline(y=0,xmin=0.0001,xmax=0.9999, color='0.05', linestyle=':', linewidth = 1)
plt.plot(rotations, Ftrace_mean, color='C2', linewidth = 1, zorder = 0, alpha=0.35)
plt.plot(rotations, np.mean(Ftrace_mean,1), linewidth = 2.5, color = 'C2')
plt.errorbar(rotations, np.mean(Ftrace_mean,1), yerr = Ftrace_sem, linewidth = 2, color = 'C2')

#plt.plot(rotations, R38G02_GC6f_Ftrace_mean, color='C0', linewidth = 0.5, zorder = 0, alpha=0.25)
#plt.plot(rotations, np.mean(R38G02_GC6f_Ftrace_mean,1), linewidth = 2, color = 'C0')
#plt.errorbar(rotations, np.mean(R38G02_GC6f_Ftrace_mean,1), yerr = stats.sem(R38G02_GC6f_Ftrace_mean,1), linewidth = 1, color = 'C0')

plt.xticks(rotations)
ax.set_xticklabels(rotations)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines['left'].set_position(('outward', 7))
ax.spines['bottom'].set_position(('outward', 7))
#for axis in ['top','bottom','left','right']:
#  ax.spines[axis].set_linewidth(0.5)
ax.set_xlabel(u'Direction [°]', fontsize = 10)
ax.set_ylabel('$\Delta F/F$', fontsize = 10)
xticks = np.array([0,90,180,270,360])
ax.set_xticklabels(xticks, size=10)
ax.set_xticks(xticks)
ax.set_yticks([-1, 0, 1, 2])
ax.set_yticklabels(('-1','0','1','2'), size=10)
plt.ylim([-1.5, 2.8])
ax.tick_params(axis='both', which='major', labelsize=10)

bbox_inches = 'tight'
#plt.savefig('C:\\Users\\gammer\\Desktop\\DATA Surface\\LPi Opponency\\plots_LPi_ms\\
#FigS2_R24E09_GC6f_DirTuning.pdf',bbox_inches='tight', dpi=600, transparent=True)

plt.show()
```
